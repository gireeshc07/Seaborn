<h1>Seaborn Python Library:</h1>
<ul>
    <li>Seaborn is a Python data visualization library built on top of Matplotlib and integrated with Pandas.</li>
    <li>It provides a high-level API for creating attractive and informative statistical graphics with ease.</li>
    <li>It is widely used in Data Analysis and Data Science because of its ability to handle complex datasets and produce aesthetically pleasing visualizations with minimal code.</li>
</ul>
<h3>Key Features:</h3>
<ul>
    <li>Built-in themes for better-looking plots.</li>
    <li>Simplifies complex visualizations like heatmaps, violin plots, and pair plots.</li>
    <li>Supports Pandas DataFrames for quick and efficient data visualization.</li>
    <li>Automatically handles statistical aggregations.</li>
</ul>

```python
# Importing Required Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

<h3>Relational Plots:</h3>
<ul>
    <li>These are used to visualize relationships b/w two or more variables.</li>
</ul>

<h4>1. Line Plot:</h4>
<ul>
    <li>It is best for time series and trends.</li>
</ul>

```python
# Load dataset
df = sns.load_dataset("flights")

# Line Plot
sns.lineplot(x="year", y="passengers", data=df)
plt.show()
```

<h4>2. Scatter Plot:</h4>
<ul>
    <li>It shows the relationship between two numerical variables.</li>
</ul>

```python
# Load dataset
df = sns.load_dataset("penguins")

# Scatter Plot
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", hue="species", size="body_mass_g", data=df)
plt.show()
```

<h3>Categorical Plots:</h3>
<ul>
    <li>These are used to visualize categorical data.</li>
</ul>

<h4>1. Bar Plot:</h4>
<ul>
    <li>It displays the mean of a numerical variable across categories.</li>
</ul>

```python
# Load dataset
df = sns.load_dataset("titanic")

# Bar Plot
sns.barplot(x="class", y="fare", hue="sex", data=df)
plt.show()
```

<h4>2. Count Plot:</h4>
<ul>
    <li>It displays count of observations per category.</li>
</ul>

```python
# Load dataset
df = sns.load_dataset("titanic")

# Count Plot
sns.countplot(x="class", hue="sex", data=df)
plt.show()
```

<h4>3. Box Plot:</h4>
<ul>
    <li>It shows distribution, quartiles, and outliers.</li>
</ul>

```python
# Loading the dataset inside the boxplot function
sns.boxplot(x="day", y="total_bill", hue="sex", data=sns.load_dataset("tips"))
plt.show()
```

<h4>4. Violin Plot:</h4>
<ul>
    <li>It combines box plot & KDE (distribution density).</li>
</ul>

```python
# Violin Plot
sns.violinplot(x="day", y="total_bill", hue="sex", data=sns.load_dataset("tips"), split=True)
plt.show()
```

<h4>5. Strip Plot:</h4>
<ul>
    <li>It displays individual data points.</li>
</ul>

```python
# Strip Plot
sns.stripplot(x="day", y="total_bill", data=sns.load_dataset("tips"), jitter=True)
plt.show()
```

<h3>Distribution Plots:</h3>
<ul>
    <li>These are used to visualize categorical data.</li>
</ul>

<h4>1. Histogram:</h4>
<ul>
    <li>Histograms show the frequency distribution of a single variable by dividing the data into bins.</li>
</ul>

```python
# Load dataset
df = sns.load_dataset("penguins")

# Histogram
sns.histplot(df["body_mass_g"], bins=30, kde=True, color="blue")
plt.show()
```

<h4>2. Kernel Density Estimation (KDE) Plot:</h4>
<ul>
    <li>A KDE plot is a smoothed version of a histogram, showing the probability density of data.</li>
</ul>

```python
# KDE Plot
sns.kdeplot(df["body_mass_g"], fill=True, color="red")
plt.show()
```

<h4>3. ECDF Plot:</h4>
<ul>
    <li>An Empirical Cumulative Distribution Function (ECDF) plot shows the proportion of data points less than or equal to a given value.</li>
</ul>

```python
# ECDF Plot
sns.ecdfplot(df["body_mass_g"])
plt.show()
```

<h4>4. Dist Plot:</h4>
<ul>
    <li>It is used to visualize the univariate distribution of a dataset.</li>
    <li>It is an old method for distribution plots and deprecated in latest versions.</li>
    <li>Instead, you should use sns.histplot() or sns.kdeplot() for modern and efficient distribution plotting.</li>
</ul>

```python
# Load dataset
df = sns.load_dataset("penguins")

# Dist Plot
sns.distplot(df["body_mass_g"], bins=30, kde=True, color="blue")  # Deprecated
plt.show()
```

<h3>Regression Plots:</h3>
<ul>
    <li>These are used to visualize distribution of data.</li>
</ul>

 <h4>1. Regression Plot:</h4>
<ul>
    <li>It draws a scatter plot with a regression line.</li>
</ul>

```python
# Load dataset
df = sns.load_dataset("tips")

# Simple Regression plot
sns.regplot(x="total_bill", y="tip", data=df)

plt.show()
```

<h4>2. Linear Regression Plot:</h4>
<ul>
    <li>A Linear Regression Plot is used when you want to create multiple regression plots grouped by a category.</li>
    <li>It is similar to sns.regplot(), but with extra faceting capabilities.</li>
</ul>

```python
# Load dataset
df = sns.load_dataset("tips")

# Simple Linear Regression Plot
sns.lmplot(x="total_bill", y="tip", data=df)

plt.show()
```

<h3>Matrix Plots:</h3>
<ul>
    <li>These are used to visualize the data in matrix form.</li>
</ul>

<h4>1. Heatmap</h4>
<ul>
    <li>A heatmap represents numerical data using colors, making it easy to spot correlations, trends, and missing values.</li>
</ul>

```python
# Load dataset
df = sns.load_dataset("flights").pivot(index="month", columns="year", values="passengers")

# Create heatmap
sns.heatmap(df, annot=True, fmt="d", cmap="coolwarm")

plt.show()

# Note: Ensure you are using the index, columns, and values parameters correctly in the pivot function. Otherwise you may get a TypeError
```

<h4>2. Cluster Map:</h4>
<ul>
    <li>A clustermap is a heatmap that groups similar data together using hierarchical clustering.</li>
</ul>

```python
# Load dataset
df = sns.load_dataset("flights").pivot(index="month", columns="year", values="passengers")

# Create cluster map
sns.clustermap(df, cmap="magma", standard_scale=1)

plt.show()
```

<h3>Multi-plot Grids:</h3>
<ul>
    <li>Multi-plot grids in Seaborn allow you to visualize multiple plots in a structured layout. They are useful for:
    <ul>
        <li>Comparing distributions across categories.</li>
        <li>Exploring relationships between multiple variables.</li>
        <li>Creating faceted (small-multiples) visualizations.</li>
    </ul>
    </li>
</ul>

<h4>1. Facet Grid:</h4>
<ul>
    <li>It creates multiple subplots based on a categorical variable. It works with scatter, line, bar, KDE, etc..</li>
</ul>

```python
# Load dataset
df = sns.load_dataset("titanic")

# Create FacetGrid
g = sns.FacetGrid(df, col="sex", row="class", margin_titles=True)

# Map a histogram onto each facet
g.map(sns.histplot, "age", bins=20, kde=True)

plt.show()

# Key Features:
'''
-> col="sex"  Creates columns for Male & Female.
-> row="class"  Creates rows for Passenger Classes.
-> margin_titles=True  Displays category labels clearly.
'''
# Insights:
'''
-> Visualizes age distribution for different genders & classes.
'''
```

<h4>2. Pair Grid</h4>
<ul>
    <li>It is used for pairwise relationships between multiple numerical variables. It is more customizable than sns.pairplot().</li>
</ul>

```python
# Load dataset
df = sns.load_dataset("penguins")

# Create PairGrid
g = sns.PairGrid(df, hue="species")

# Scatter plot for upper triangle
g.map_upper(sns.scatterplot)

# KDE plot for lower triangle
g.map_lower(sns.kdeplot, fill=True)

# Histogram for diagonal
g.map_diag(sns.histplot, kde=True)

plt.show()

# Key Features:
'''
-> map_upper(): Upper triangle scatter plot.
-> map_lower(): Lower triangle KDE plot.
-> map_diag(): Diagonal histogram with KDE.
-> hue="species": Color-codes species.
'''

# Insights:
'''
-> How variables relate to each other (scatter).
-> Distribution of each variable (diagonal histograms).
-> Density of observations (KDE).
'''
```

<h4>3. Pair Plot:</h4>
<ul>
    <li>It is a quick version of sns.PairGrid(). It creates scatter & histogram plots for all numerical variables.</li>
</ul>

```python
# Load dataset
df = sns.load_dataset("penguins")

# Pair Plot
sns.pairplot(df, hue="species", diag_kind="kde")
plt.show()

# Best for:
'''
-> Exploring multiple numerical features quickly.
-> Finding clusters & correlations.
'''

# Insights:
'''
-> Diagonal shows distribution (diag_kind="kde").
-> Lower triangle shows scatter plots.
'''
```

<h4>4. Joint Plot:</h4>
<ul>
    <li>It creates scatter + histogram or KDE in one figure. It is used for bivariate (two-variable) analysis.</li>
</ul>

```python
# Load dataset
df = sns.load_dataset("penguins")

# Join Plot
sns.jointplot(data=df, x="bill_length_mm", y="bill_depth_mm", kind="scatter", hue="species")
plt.show()

# Key Features:
'''
-> kind="scatter"  Scatter plot (default).
-> kind="kde"  KDE contour plot.
-> kind="hex"  Hexbin plot for dense data.
'''

# Insights:
'''
-> Shows relationship between two variables.
-> Reveals density distribution.
'''
```
